```python
df = pd.read_csv('antropometria-dataset.csv') 
df
```

```python
sns.boxplot(data=df.loc[:,['talla']])
```

```python
tabla = pd.pivot_table(df,
                       index = ['desc_ent'],
                       columns = ['sexo'],
                       values = ['peso'],
                       aggfunc = {'peso':['median']}
                      )
tabla
```

```python
tablados = pd.pivot_table(df, 
                          index = ['desc_ent'],
                          columns = ['sexo'],
                          values = ['cadera'],
                          aggfunc = {'cadera':['median']}
                         )
tablados
                          

```

```python
line_plot = tablados[('cadera','median',2)].plot(kind='line')
```

```python
tablatres = pd.pivot_table(df,
                           index = ['desc_ent'],
                           columns = ['sexo'],
                           values = ['ropa'],
                           aggfunc = {'ropa':['sum']}
                          )
tablatres
```

```python
by_ropa = tablatres.sort_values(('ropa','sum',2),ascending=False) 
top10 = by_ropa.head(10)
top10
```

```python
top10[('ropa','sum',2)].plot(kind = 'barh',
                             figsize = (20,10),
                             color = 'y',
                             alpha = 0.5)
```
