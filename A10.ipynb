# Sobre el data set
## Las enfermedades cardiovasculares son la causa principal de muerte en el mundo, quitándole la vida a alrededor de 18 millones de personas al año lo cual es el 31% de todas las muertes

Este set de datos contiene 12 “features” que indican si el paciente tiene ciertas características relacionadas a la salud y una de ellas nos indica si el paciente murió a causa de un fallo en el corazón.
La mayoría de las enfermedades cardiovasculares se pueden prevenir tomando en cuenta las conductas que ponen en riesgo este sistema. Estos factores incluyen el consumo de tabaco, dieta no saludable, obesidad, inactividad física, consumo excesivo de alcohol etc. 


# Variables
Age= Edad
Anemia= si tiene anemia o no,
Creatine= nivel de CPK en la sangre,
Diabetes= si el paciente tiene diabetes,
ejection fraction= porcentaje de sangre que deja los pulmones en cada contraction,
high blood pressure= si el paciente tiene hypertension,
platelets= cantidad de plaquetas en la sangre,
serum_sodium=nivel de serum sodium en la sangre,
Sex= sexo del paciente,
smoking= si el paciente fuma o no ,
time= periodo de seguimiento en dias,
death_event= si el paciente fallecio o no,

## 8. Cual es el modelo de regression lineal entre la variable dependiente DEATH_EVENT 
## y las variables independientes que mas correlacion tienen como age, high_blood_pressure, serum_creatinine, serum_sodium, ejection_fraction

```python
pip install statsmodels
```

```python
import statsmodels.api as sm
```

```python
model = sm.OLS.from_formula ('DEATH_EVENT ~ age+high_blood_pressure+serum_creatinine+serum_sodium+ejection_fraction', data = base)
result = model.fit()
result.summary()
```
